## Easy (+0.1)

```python
import pandas as pd
import seaborn as sns
```

```python
def complains_age(file: str) -> dict:
    res = dict()
    df = pd.read_csv(file, sep=';')
    res['complains'] = df['Complain'].sum()
    res['avg'] = 2024 - df['Year_Birth'].mean()
    df = df[df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(
        axis=1) > 0]
    df = df[['Year_Birth', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
             'MntGoldProds']]
    res['avg_mnt'] = 2024 - df['Year_Birth'].mean()
    return res
```

```python
ans = complains_age('../data/marketing_campaign.csv')
complains = ans['complains']
avg = ans['avg']
avg_mnt = ans['avg_mnt']
```

```python
f'{complains} - Количество жалоб'
```

```python
f'{avg} - Средний возраст всех людей в датасете'
```

```python
f'{avg_mnt} - Средний возраст покупателей'
```

Средний возраст покупателей не отличается от среднего возраста всех людей в датасете

## Medium (+0.4)

```python
df = pd.read_csv('../data/marketing_campaign.csv', sep=';')

df['Education'] = df['Education'].astype('category')
df['Marital_Status'] = df['Marital_Status'].astype('category')

df['Age'] = 2024 - df['Year_Birth']
sns.pairplot(df, vars=['Education', 'Marital_Status', 'Income', 'Kidhome', 'Teenhome', 'Recency', 'NumWebPurchases', 'NumStorePurchases', 'Age'])
```

Заметим, что годовой доход семьи клиента влияет на количество покупок в магазине и на сайте компании (Income / NumStorePurchases, NumWebPurchases) 

Также наличие отношений и их "степень" влияет на количество детей в семье клиента (Marital / Teenhome )
